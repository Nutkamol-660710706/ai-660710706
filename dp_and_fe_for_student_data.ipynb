```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df_student = pd.read_csv("/content/sample_data/student.csv")
df_student
```

1.เมื่อพบค่า NaN ในชื่อ จะเปลี่ยนให้เป็น ไม่พบ เพื่อไม่ให้เป็นค่าว่าง

```python
df_student['ชื่อ'].fillna('ไม่ระบุ', inplace=True)
```

```python
df_student
```

2.แทนที่ข้อมูลที่มีอายุผิดปกติ ด้วยค่าเฉลี่ยของอายุไปแทน

```python
Q1 = df_student['อายุ'].quantile(0.25)
Q3 = df_student['อายุ'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_student.loc[df_student['อายุ'] < lower_bound, 'อายุ'] = np.nan
df_student.loc[df_student['อายุ'] > upper_bound, 'อายุ'] = np.nan

mean_age = np.round(df_student['อายุ'].mean(), 2)
df_student['อายุ'].fillna(mean_age, inplace=True)
```

```python
df_student
```

3.ปรับเพศให้แสดงแบบเดียวกันคือ ชาย และ หญิง

```python
def standardize_Gender (Gender):
    if Gender in ['ชาย', 'ช', 'M', 'Male']:
        return 'ชาย'
    elif Gender in ['หญิง', 'ญ', 'F','Female']:
        return 'หญิง'
    else:
        return Gender
```

```python
df_student['เพศ'] = df_student['เพศ'].apply(standardize_Gender)
```

```python
df_student
```

4.นำค่าฐานนิยมของเพศไปแทนในส่วนของเพศที่เป็นค่าว่าง

```python
df_student['เพศ'].fillna(df_student['เพศ'].mode()[0], inplace=True)
```

```python
df_student
```

5.ปรับเพศให้แสดงแบบเดียวกันคือ วิทยาศาสตร์ , ศิลปศาสตร์, วิศวกรรมศาสตร์

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวกรรมศาสตร์', 'วิศวะ', 'วิศวกรรม' , 'วิศวฯ']:
        return 'วิศวกรรมศาสตร์'
    else:
        return name
```

```python
df_student['คณะ'] = df_student['คณะ'].apply(standardize_faculty)
```

```python
df_student
```

6.นำค่าฐานนิยมของคณะไปแทนในค่าว่าง

```python
df_student['คณะ'].fillna(df_student['คณะ'].mode()[0], inplace=True)
```

```python
df_student
```

7.ทำค่า GPA ที่ผิดปกติให้เป็นค่าว่างแล้ว นำค่าเฉลี่ยของ GPA ไปแทนในค่าว่าง

```python
df_student.loc[df_student['GPA'] < 0, 'GPA'] = np.nan
df_student.loc[df_student['GPA'] > 4, 'GPA'] = 4.0
mean_gpa = np.round(df_student['GPA'].mean(), 2)
df_student['GPA'].fillna(mean_gpa, inplace=True)
```

```python
df_student
```

8.แทนค่าว่างในน้ำหนักด้วยค่าเฉลี่ยของ น้ำหนัก

```python
mean_age = np.round(df_student['น้ำหนัก_kg'].mean(), 2)
df_student['น้ำหนัก_kg'].fillna(mean_age, inplace=True)
```

```python
df_student
```
